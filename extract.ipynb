```python
import investpy
import pandas as pd


# 브라질 국채 목록 조회
brazil_bonds = investpy.get_bonds(country='Brazil')
brazil_bonds
```

```python
bond_data = investpy.get_bond_historical_data(bond='Brazil 3M', from_date='11/10/2024', to_date='11/04/2025')
bond_data
```

```python
bond_data.to_csv('Brazil-6M.csv')
```

```python
import yfinance as yf

usd_brl_futures = yf.Ticker("BRL=X")

usd_brl = usd_brl_futures.history(period="6mo")
usd_brl = usd_brl.loc['2024-10-11':'2025-04-11', :]
brl_usd = 1 / usd_brl
```

```python
brl_usd.index = brl_usd.index.strftime('%Y-%m-%d')
brl_usd.index = pd.to_datetime(brl_usd.index)
brl_usd
```

```python
brl_usd.to_csv('brl_usd.csv')
```

```python
import yfinance as yf

usd_krw_futures = yf.Ticker("KRW=X")

usd_krw = usd_krw_futures.history(period="6mo")
usd_krw = usd_krw.loc['2024-10-11':'2025-04-11']
krw_usd = 1 / usd_krw
```
