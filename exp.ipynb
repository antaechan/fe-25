```python
import investpy
import pandas as pd

# 브라질 국채 목록 조회
brazil_bonds = investpy.get_bonds(country='Brazil')
brazil_bonds
```

```python
bond_data = investpy.get_bond_historical_data(bond='Brazil 6M', from_date='01/01/2023', to_date='28/03/2025')
bond_data
```

```python
bond_data.to_csv('Brazil-6M.csv')
```

```python
import yfinance as yf

brl_usd_futures = yf.Ticker("BRL=X")

brl_usd_data = brl_usd_futures.history(period="1y")
brl_usd_data.tail()
```

```python
brl_usd_data.index = brl_usd_data.index.tz_convert("UTC").tz_localize(None)
brl_usd_data
```

```python
brl_usd_data.to_csv('brl_usd.csv')
```
